```
from mpl_toolkits import mplot3d
```

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(x, y):
    return (x**2 * y**2)
  
x = np.linspace(-5, 5, 40)
y = np.linspace(-5, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

ax.view_init(20, 185)

ax.scatter(0, 0, color="blue")
```

```
def f(x, y):
    return (-y**2)
  
x = np.linspace(-5, 5, 40)
y = np.linspace(-5, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

ax.view_init(10, 140)

ax.scatter(0, 0, color="red")
```

```
def f(x, y):
    return (x*y)
  
x = np.linspace(-5, 5, 40)
y = np.linspace(-5, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

ax.view_init(24, 165)

# Gradient = 0 at (0, 0, 0).
ax.scatter(0, 0, color="green")
```
